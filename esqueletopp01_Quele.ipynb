```python
import numpy as np
import pandas as pd

# === CONFIGURAÇÕES ===
np.random.seed(42)
ETA = 0.01         # Taxa de aprendizado
MAX_EPOCAS = 100   # Número máximo de épocas por treinamento
NUM_TREINAMENTOS = 5  # Quantidade de repetições com inicializações diferentes

# === 1. CARREGAR DADOS ===
# TODO: Carregar os arquivos CSV de treino e validação

df_train = pd.read_excel("/content/TabelaSeção36RNA.xlsx")
df_val   = pd.read_excel("/content/Tabela33RNA.xlsx")

# TODO: Extrair as features (x1, x2, x3) e o rótulo alvo (d) como arrays numpy
X_train = df_train[['x1', 'x2', 'x3']].values
d_train = df_train['d'].values
X_val   = df_val[['x1', 'x2', 'x3']].values
#d_val   = df_val['d'].values

# === 2. ADICIONAR TERMOS DE BIAS ===
# TODO: Inserir uma coluna de 1s à esquerda das matrizes de features
# Dica: use np.column_stack([np.ones(...), X])
X_train_bias = np.column_stack([np.ones(X_train.shape[0]), X_train])
X_val_bias   = np.column_stack([np.ones(X_val.shape[0]), X_val])

# === 3. FUNÇÃO DE ATIVAÇÃO (DEGRAU BIPOLAR) ===
def sgn(z):
    # TODO: Retornar 1 se z >= 0, caso contrário retornar -1
    return 1 if z >= 0 else -1

# Lista para armazenar os resultados de cada repetição
resultados_treinamentos = []

# === 4. LOOP DE TREINAMENTOS ===
for i in range(NUM_TREINAMENTOS):
    # Muda a semente para gerar pesos iniciais diferentes a cada execução
    np.random.seed(123 + i)

    # TODO: Inicializar o vetor de pesos com 4 valores aleatórios entre 0 e 1
    w = np.random.rand(4)
    w_inicial = w.copy()

    epoca = 0
    convergiu = False

    while epoca < MAX_EPOCAS and not convergiu:
        erros = 0
        for j in range(len(X_train_bias)):
            x = X_train_bias[j]
            d = d_train[j]

            # TODO: Calcular a saída líquida (produto escalar entre pesos e entrada)
            z = np.dot(w, x)

            # TODO: Aplicar a função de ativação
            y = sgn(z)

            # TODO: Regra de aprendizado do Perceptron
            # Se a previsão estiver errada, ajuste os pesos
            if y != d:
                # Dica: w = w + ETA * (d - y) * x   (ou a variante vista em aula)
                w = w + ETA * (d - y) * x
                erros += 1

        epoca += 1

        # TODO: Se não houve erros na época inteira, marque como convergido
        if erros == 0:
            convergiu = True

    # TODO: Salvar os resultados deste treinamento
    resultados_treinamentos.append({
        'w_inicial': w_inicial,
        'w_final': w,
        'epocas': epoca if convergiu else MAX_EPOCAS  # Se convergiu, use 'epoca'. Caso contrário, use 'MAX_EPOCAS'
    })

# === 5. EXIBIR TABELA 3.2 (Resultados do Treino) ===
print("\n" + "="*80)
print("TABELA 3.2 – Resultados dos treinamentos do Perceptron")
print("="*80)
print(f"{'Trein.':<8} {'w0_i':<8} {'w1_i':<8} {'w2_i':<8} {'w3_i':<8} "
      f"{'w0_f':<8} {'w1_f':<8} {'w2_f':<8} {'w3_f':<8} {'Épocas':<8}")
print("-" * 80)

for i, res in enumerate(resultados_treinamentos):
    wi = res['w_inicial']
    wf = res['w_final']
    epocas = res['epocas']

    # TODO: Complete o f-string abaixo para exibir os valores formatados
    print(f"{i+1}º(T{i+1})  {wi[0]:<8.3f} {wi[1]:<8.3f} {wi[2]:<8.3f} {wi[3]:<8.3f} "
          f"{wf[0]:<8.3f} {wf[1]:<8.3f} {wf[2]:<8.3f} {wf[3]:<8.3f} {epocas:<8}")

# === 6. EXIBIR TABELA 3.3 (Validação) ===
print("\n" + "="*80)
print("TABELA 3.3 – Amostras de óleo para validar a rede Perceptron")
print("="*80)
print(f"{'Amostra':<8} {'x1':<9} {'x2':<9} {'x3':<9} {'y(T1)':<6} {'y(T2)':<6} {'y(T3)':<6} {'y(T4)':<6} {'y(T5)':<6}")
print("-" * 80)

for idx in range(len(X_val_bias)):
    x = X_val_bias[idx]
    # TODO: Exibir os valores de x1, x2, x3 da amostra atual
    print(f"{idx+1:<8} {x[1]:<9.4f} {x[2]:<9.4f} {x[3]:<9.4f}", end=" ")

    # TODO: Para cada modelo treinado, calcule a previsão na amostra de validação
    for i in range(NUM_TREINAMENTOS):
        w = resultados_treinamentos[i]['w_final']
        # Dica: calcule z = np.dot(w, x) e depois y = sgn(z)
        y = sgn(np.dot(w, x))
        print(f"{y:<6}", end="")
    print()  # Pula linha após cada amostra
```

## Respostas às Questões Teóricas

### Questão 4 — Por que o número de épocas varia entre treinamentos?

O número de épocas varia porque os pesos iniciais são definidos especificamente no início de cada treinamento. Como cada treinamento parte de um ponto diferente no espaço de pesos, o caminho percorrido até encontrar uma solução também é diferente, o que naturalmente leva a detalhes de épocas distintas.
No nosso caso todos os treinamentos chegaram em 100 épocas porque o algoritmo não convergiu em nenhum deles, o que já adianta a resposta da próxima questão.

### Questão 5 — As classes são linearmente separáveis?

Não é possível afirmar que as classes sejam linearmente separáveis. A prova disso está nos próprios resultados: nenhum dos cinco treinamentos convergiu, todos atingiram o limite máximo de 100 épocas sem zerar os erros.
Isso é relevante porque o Teorema da Convergência do Perceptron garante que, se as classes forem linearmente separáveis, o algoritmo converge necessariamente em um número finito de épocas. Como isso não aconteceu em nenhum dos cinco treinamentos com inicializações diferentes, podemos concluir que não existe um hiperplano no espaço (x1, x2, x3) capaz de separar perfeitamente as classes P1 e P2, ou seja, os dados não são linearmente separáveis.
