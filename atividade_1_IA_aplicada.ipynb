```
import numpy as np
import matplotlib.pyplot as plt
```

```
# ============================================================
# EXERCÍCIO 1 - Potencial de Ativação - Neurônio Único (n=3)
# ============================================================

X = [1, 2, 3]
W = [0.2, 0.8, -0.5]
bias = 2

u = sum(w * x for w, x in zip(W, X)) + bias
print(f"Exercício 1 | Saída u = {u}")
```

```
# ============================================================
# EXERCÍCIO 2 - Potencial de Ativação - Neurônio Único (n=4)
# ============================================================

X = [1, 2, 3, 2.5]
W = [0.2, 0.8, -0.5, 1.0]
bias = 2

u = sum(w * x for w, x in zip(W, X)) + bias
print(f"Exercício 2 | Saída u = {u}")
```

```
# ============================================================
# EXERCÍCIO 3 - Potencial de Ativação - Camada de Neurônios
# ============================================================

X   = [1, 2, 3, 2.5]
W1  = [0.2,  0.8,  -0.5,  1.0]
W2  = [0.5, -0.91,  0.26, -0.5]
W3  = [-0.26, -0.27, 0.17, 0.87]
bias1, bias2, bias3 = 2, 3, 0.5

u1 = sum(w * x for w, x in zip(W1, X)) + bias1
u2 = sum(w * x for w, x in zip(W2, X)) + bias2
u3 = sum(w * x for w, x in zip(W3, X)) + bias3

print(f"Exercício 3 | Saídas u = [{round(u1,4)}, {round(u2,4)}, {round(u3,4)}]")
```

```
# ============================================================
# EXERCÍCIO 4 - Função de Ativação: Bipolar Aproximada
# ============================================================

def bipolar_aprox(x):
    return 1 if x >= 0 else -1

x_vals = np.arange(-5.0, 5.1, 0.1)
y_vals = np.array([bipolar_aprox(xi) for xi in x_vals])

plt.figure()
plt.plot(x_vals, y_vals, 'b', label='f(x) bipolar aprox.')
plt.scatter([0], [bipolar_aprox(0)], color='red', zorder=5, label='f(0) = 1')
plt.title('Função Bipolar Aproximada')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('bipolar_aprox.png', dpi=300, bbox_inches='tight')
plt.show()
```

```

# ============================================================
# EXERCÍCIO 5 - Função de Ativação: Rampa Simétrica
# ============================================================

def rampa_simetrica(x):
    if x < -1:
        return -1
    elif x > 1:
        return 1
    else:
        return x

x_vals = np.arange(-5.0, 5.1, 0.1)
y_vals = np.array([rampa_simetrica(xi) for xi in x_vals])

plt.figure()
plt.plot(x_vals, y_vals, 'r', label='f(x) = rampa simétrica(x)')
plt.scatter([-1], [rampa_simetrica(-1)], color='green', zorder=5, label='f(-1) = -1... (0)')
plt.scatter([ 1], [rampa_simetrica( 1)], color='green', zorder=5, label='f(1) = 1')
plt.title('Função Rampa Simétrica')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('rampa_simetrica.png', dpi=300, bbox_inches='tight')
plt.show()
```

```
# ============================================================
# EXERCÍCIO 6 - Função de Ativação: Tangente Hiperbólica
# ============================================================

def tangente_hiperbolica(x, b=1.0):
    return (1 - np.exp(-b * x)) / (1 + np.exp(-b * x))

x_vals = np.arange(-5.0, 5.1, 0.1)

plt.figure()
for b in [0.5, 1.0, 2.0, 5.0]:
    y_vals = tangente_hiperbolica(x_vals, b)
    plt.plot(x_vals, y_vals, label=f'b = {b}')

plt.title('Função Tangente Hiperbólica')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('tangente_hiperbolica.png', dpi=300, bbox_inches='tight')
plt.show()
```

```
# ============================================================
# EXERCÍCIO 7 - Função de Ativação: Gaussiana
# ============================================================

def gaussiana(x, sigma=1.0):
    return np.exp(-x**2 / (2 * sigma**2))

x_vals = np.arange(-5.0, 5.1, 0.1)

plt.figure()
for sigma in [0.5, 1.0, 2.0, 3.0]:
    y_vals = gaussiana(x_vals, sigma)
    plt.plot(x_vals, y_vals, label=f'σ = {sigma}')

plt.title('Função Gaussiana')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('gaussiana.png', dpi=300, bbox_inches='tight')
plt.show()
```

```
# ============================================================
# EXERCÍCIO 8 - Função de Ativação: Linear
# ============================================================

def linear(x):
    return x

x_vals = np.arange(-5.0, 5.1, 0.1)
y_vals = linear(x_vals)

plt.figure()
plt.plot(x_vals, y_vals, 'b', label='f(x) = x')
plt.title('Função de Ativação Linear')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig('linear.png', dpi=300, bbox_inches='tight')
plt.show()
```
